### TARGET값 만들기

target 기준: 최근 3개월 소비가 1번이라도 있고, 최근 1개월 소비가 있었다면 >> 꾸준히 돈을 소비하고 있다.  
꾸준히 돈을 소비하므로 고객의 만족도가 높다고 간주.

```python
cust_df['TARGET'] = (
    (cust_df['imp_op_var39_comer_ult3'] > 0) &
    (cust_df['imp_op_var39_comer_ult1'] > 0) &
    (cust_df['imp_op_var39_comer_ult1'] / cust_df['imp_op_var39_comer_ult3'] >= 0.3)
).astype(int)
```

```python
print(cust_df['TARGET'].value_counts())
unsatisfied_cnt=cust_df[cust_df['TARGET']==1].TARGET.count()
total_cnt=cust_df.TARGET.count()
print('unsatisfied rate: {0:.2f}'.format((unsatisfied_cnt/total_cnt)))
```
